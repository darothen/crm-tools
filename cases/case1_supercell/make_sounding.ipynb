Read in the sounding

```python
P0    = 1000.0
P_sfc = 887.2000 

sounding = pd.read_table("kuhlman2006.snd", sep="\s*")
print sounding.head()
theta_sfc = sounding.theta.ix[0]
qv_sfc = sounding.qv.ix[0]
T_sfc = theta_sfc/((P0/P_sfc)**(Rd/cp))
print "\n T_sfc = ", T_sfc
print sounding
```

Apply the hydrostatic approximation to estimate the pressure profile coincident with the thermodynamic profile. Then, plot the sounding on a Skew-T/Log-P plot.

```python
fig = figure()
ax = fig.add_subplot(111)
ax.plot(sounding['theta'], sounding['z'])
heights = sounding['z']

## hydrostatic pressure
thetavs = sounding["theta"].values*(1. + 0.61*sounding["qv"]/1000.)
zs = sounding['z'].values
gamma = Rd/cp
X = g/Rd/((P0*100.)**(-1.*gamma))
new_ps = [P_sfc*100., ]
for i in xrange(len(zs)-1):
    z_ip1, z_i = zs[i+1], zs[i]
    dz = z_ip1 - z_i
    ave_inv_thetav = 0.5*(1./thetavs[i] + 1./thetavs[i+1])
    p_next = new_ps[i]**(gamma) - gamma*X*ave_inv_thetav*dz
    p_next = p_next**(1./gamma)
    new_ps.append(p_next)
    #print i, z_i, z_ip1, new_ps[-1], p_next
#new_ps.append(new_ps[-1]-(zs[-1]-zs[-2])*1.3*9.8)
new_ps = np.array(new_ps)
sounding['pressure'] = pd.Series(new_ps/100., index=heights.index)

## Temperatures
thetas = sounding['theta']
#print len(thetas), len(new_ps)
temps = thetas/((P0/(new_ps/100.))**0.286)
sounding['temperature'] = pd.Series(temps, index=heights.index)

## Dewpoints
dewpoints = calc_dewpoint(sounding['temperature']-273.15, 
                          sounding['pressure'],
                          sounding['qv']/1000.)
sounding['dewpoint'] = pd.Series(dewpoints, index=heights.index)

fig, [height_ax, temp_ax] = subplots(1, 2, figsize=(24, 10), sharey=True)
height_ax.plot(sounding['pressure'], heights, marker='o')
height_ax.set_xlim(50, 1000)
#height_ax.set_ylim(0, np.max([left['height'].max(), right['height'].max()]))
height_ax.set_ylim(0, 20000)
height_ax.set_xlabel("Pressure (hPa)")
height_ax.set_ylabel("Height (m)")
height_ax.legend()

temp_ax.plot(sounding['temperature'], heights, marker='o')

## Set-up sounding for plotting
sounding['temperature'] = temps - 273.15
sounding['height'] = sounding['z']
sounding['mixing_ratio'] = sounding['qv']
sounding['us'] = sounding['U']/0.51444
sounding['vs'] = sounding['V']/0.51444
```

```python
t_str = "NCAR MGLASS - 2022 UTC\n29 June 2000\nIdealized (Kuhlman, 2006)"
plot_skewt(sounding, title=t_str,
           lift_parcel=True, plot_winds=True, diags=["CAPE", ],
           label_altitudes=True)
savefig("sounding.pdf")
savefig("sounding.png")
```

Interpolate the temperature, moisture, and pressure profiles to 
a 50-layer grid with $\Delta z = 500$m

```python
from scipy.interpolate import interp1d
#zs_grid = np.arange(125., 2e4, 250.)
#zs_grid = np.arange(0., 2e4, 100.)
#zs_grid = sounding['height'].values

dz = 250.
nz = 76
zs_grid = np.arange(0., (dz*nz)+dz/10., dz)
print zs_grid

xs = sounding['height'].values

def interp_to_grid(xs, ys, zs=zs_grid):
    f = interp1d(xs, ys, bounds_error=False, fill_value=ys[-1])
    return f(zs)

## Temperature
new_temp = interp_to_grid(xs, sounding['temperature'].values)

## Water Vapor
new_qv = interp_to_grid(xs, sounding['mixing_ratio'].values)

## Pressure
new_pressure = interp_to_grid(xs, sounding['pressure'].values)

## Winds
new_u = interp_to_grid(xs, sounding['U'].values)
new_v = interp_to_grid(xs, sounding['V'].values)

## Dewpoint
new_dewpoint = calc_dewpoint(new_temp, 
                             new_pressure,
                             new_qv/1000.)

## Set-up sounding for plotting
new_sounding = pd.DataFrame(index=np.arange(len(zs_grid)))

new_sounding['pressure'] = new_pressure
new_sounding['temperature'] = new_temp
new_sounding['height'] = zs_grid
new_sounding['dewpoint'] = new_dewpoint
new_sounding['mixing_ratio'] = new_qv
new_sounding['us'] = new_u/0.51444
new_sounding['vs'] = new_v/0.51444
```

```python
t_str = "NCAR MGLASS - 2022 UTC\n29 June 2000\nInterpolated (Kuhlman, 2006)"
plot_skewt(new_sounding, title=t_str,
           lift_parcel=True, plot_winds=True, diags=["CAPE", ],
           label_altitudes=True)
savefig("sounding_interp.pdf")

## Immediately convert the sounding velocities back to m/s from knots!
new_sounding['us'] = new_sounding['us']*0.51444
new_sounding['vs'] = new_sounding['vs']*0.51444
```

Check for hydrostatic balance

```python
from soundings import Rd

tvs = ((new_sounding['temperature']+273.15)*\
       (1. + 0.61*new_sounding['mixing_ratio']/1000.))
ps = new_sounding['pressure']*100.
rho = ps/(Rd*tvs)

###

def d_dz(phis, xs):
    phi_im1, phi_i, phi_ip1 = phis
    x_im1, x_i, x_ip1 = xs
    
    dx1 = x_i - x_im1
    dx2 = x_ip1 - x_i
    
    num = (dx1**2)*phi_ip1 + (dx2**2 - dx1**2)*phi_i - (dx2**2)*phi_im1
    den = dx1*dx2*(dx1 + dx2)
    return num/den

dps_dz = np.zeros_like(ps)
zs = new_sounding['height']
for k in xrange(1, len(ps)-1):
    dps_dz[k] = d_dz([ps[k-1], ps[k], ps[k+1]], 
                     [zs[k-1], zs[k], zs[k+1]])

print pd.DataFrame({'dp_dz': dps_dz, 
                    '-rho*g': -rho*9.8})
```
