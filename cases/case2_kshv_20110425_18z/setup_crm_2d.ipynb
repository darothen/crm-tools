## Setup CRM - 2D Version

---

This notebook is designed to quickly setup the initial conditions necessary to run the CRM. It will create the namelist files and some basic analysis plots of the soundings used to initialize the model. Then, it will copy the files in place as well as to the model run directory. The following set of cells will automate the setup and visual inspection of initial conditions for Chien's cloud-resolving model. For now, it is assumed that a homogeneous profile will be applied to the entire model domain.

__2D CRM__

```python
CASENAME = "kshv"
CASEDIR = "."

## Domain and sounding info
nx, ny, nz = 400, 1, 50
sounding_filename = "kshv_20110425_18z.snd"

namelist = {
    ## Case settings
    "casename"      : CASENAME,
    #"homedir"       : "/Users/daniel/workspace/crm71_dev",
    "file_init"     : '%s_init' % CASENAME,
    "file_rktbl"    : 'rktable32.dat',
    "file_cheminit" : 'CHEMINIT_32_KHSV_04252011_18Z', 
            
    ## Time steps
    "dt"   : 1.0, # basic model time step (seconds)
    "dtau" : 0.1, # small time step for sound waves (seconds)
    "ds"   : 2.0, # time step for xxx (seconds)
    
    ## Resume control
    "nresume" : 0, # 0 = new run with one time field init data;
                   # 1 = start a run with two time field init data/apply disturbance
                   # 2 = resume a run with two time field init data
    
    ## Start point in step
    "nstart" : 1,
    
    ## Start time in 24 hour base:
    # The main program will convert to seconds, e.g.
    # start_in_s = 12.0 -> 43200 s (12 hours -> 43200 s)
    "start_in_s" : 12.0,
    
    ## Integration length:
    "nstop" : 8, # 24 if nresume == 0; 120 when nresume == 1
    "nunit" : 3600, # 3600: hour
                    # 60: minute
                    # dt: step
    
    ## Frequency of output data in step:
    "iax" : -2,
    
    ## Radiation calculation frequency:
    # example: 60 would be every 5 minutes if dt = 5 seconds
    # set to negative int or 0 to close radiation call
    "iradx" : 100, 
    
    ## Chemistry Package
    #"if_chem" : 1, # obsolete
    # Rxn calculation frequency in steps
    # (e.g. 6 would be every 30 seconds in 5 sec time step)
    "i_creact" : 6,
    
    ## Dimensions
    "ndims" : 2,   # numbe2r of dimensions
    "dx" : 2000.0, # x-width in meters
    "dy" : 2000.0, # y-width in meters
    "dz" : 500.0,  # height in meters
    
    ## Thickness of top damping layer
    # normally can be set to 5000(m)/dz(m)
    "kdamp" : 15,
    
    ## LBC parameters
    "eps" : 0.0, # relaxation coefficient
    "if_lbcdamp" : 0, # if > 0, tau = dx/C = n in seconds
    #"if_plbc" : 0, # periodic lbc option; obsolete
    
    ## Initial wind adjustments
    # Also controls the maximum 
    "u0shift" : 0.0, # m/s
    "v0shift" : 0.0, # m/s
    
    ## Minimum K for wind/scalar
    # reference: fkmin_u = 25.0, fkmin_s = 3.*fkmin_u
    "fkmin_u" : 25.0,
    "fkmin_s" : 75.0,
    
    ## Warm cloud?
    "nwarm" : 0, # 1 = warm, 0 = cold
    
    ## Go forward one step after smoothing?
    "nsmooth" : 1, # 1 = go one step forward, 0 = don't
    
    ## CCN and IN parameters
    "xnc0_c"  : 100.0e6,  # CCN nucleation constant C in CCN 1/kg
    "xnc0_k"  : 0.5,     # CCN nucleation constant k (0.7 for maritime, 0.5 for cont)
    "xn_ccn0" : 250.0e6, # sfc or max CCN concentration (1/kg)
    "xn_in0"  : 100.0e3, # sfc or max IN concentration (1/kg)
    
    ## Filter
    "if_filter" : -5, # if > 0, apply filter per n(dt) steps, e.g. 12
    
    ## Solar radiation
    "j_day" : 115, # day of the year from 1 to 365
    "ctr_lat" : 32.44, # latitude at center of model domain, from -90 to 90
    
    ## Initial perturbation
    "nbulb" : 1, # 1 = use bulb, 0 = don't
    "dptil" : 3.0, # max init perturbation of temperature in degree
    "dqv" : 0.9, # max init perturbation of water vapor mixing ratio in %-age
    "mx"  : 50,
    "mx1" : 5,
    "mx2" : 5,
    "ioe" : 2,
    "my"  : 100,
    "my1" : 5,
    "my2" : 5,
    "joe" : 2,
    "mz"  : 3,
    "mz1" : 3,
    "mz2" : 3,
    "koe" : 2,
}
```

```python
## Write namelist

nl_order = ["dt", "dtau", "ds", "nresume", "nstart", "start_in_s", "nstop",
            "nunit", "iax", "iradx", "i_creact", "ndims", "dx", "dy", "dz",
            "kdamp", "eps", "if_lbcdamp", "u0shift", "v0shift", "fkmin_u",
            "fkmin_s", "nwarm", "nsmooth", "xnc0_c", "xnc0_k", "xn_ccn0",
            "xn_in0", "if_filter", "nbulb", "dptil", "dqv", "mx", "mx1", 
            "mx2", "ioe", "my", "my1", "my2", "joe", "mz", "mz1", "mz2",
            "koe", "j_day", "ctr_lat", "casename", "file_init",
            "file_rktbl", "file_cheminit"]

namelist_fn = "cm%1dd.nml" % namelist['ndims']
with open(namelist_fn, "w") as f:
    f.write("$cm_in \n")
    for key in nl_order:
        value = namelist[key]
        f.write("   %-13s = %r \n" % (key, value))
    f.write("$END \n")
print namelist_fn
```

####Thermodynamic soundings/profiles

```python
dx, dy, dz = namelist['dx'], namelist['dy'], namelist['dz']
sounding = soundings.read_sounding(sounding_filename)

dirs = np.pi*sounding['direction'].values/180. # in radians
us = -sounding['sknt'].values*np.sin(dirs)
vs = -sounding['sknt'].values*np.cos(dirs)
sounding['us'] = pd.Series(us, index=sounding.index)
sounding['vs'] = pd.Series(vs, index=sounding.index)
sounding = sounding.dropna()
fig, ax1 = subplots(1,1, figsize=(7, 7), 
                           subplot_kw={'axisbg': 'w'} )
sounding['qv'] = sounding['mixing_ratio']
soundings.plot_skewt(sounding, ax=ax1, lift_parcel=True, plot_winds=True,
                   diags=['CAPE'], title="KSHV\n2011-04-25 18:00:00Z",
                   T_color='grey', Td_color='grey', parcel_color='grey',
                   barb_loc=42., barb_color='grey',
                   T_left=-32.)

zs = np.linspace(0, dz*nz, nz+1)
from scipy.interpolate import interp1d

def prepend(val, l):
    new_list = [val,]
    new_list.extend(l)
    return np.array(new_list)

heights = sounding['height']

model_temperature = interp1d(heights, 
                             sounding['temperature'], kind='cubic')(zs[1:])
model_temperature = prepend(sounding['temperature'].values[0], model_temperature)
model_temperature = soundings.filter_121(model_temperature, 1)

model_pressure = interp1d(heights, 
                          sounding['pressure'], kind='slinear')(zs[1:])
model_pressure = prepend(sounding['pressure'].values[0], model_pressure)

#model_dewpoint = interp1d(heights, 
#                          sounding['dewpoint'], kind='cubic')(zs[1:])
#model_dewpoint = prepend(sounding['dewpoint'].values[0], model_dewpoint)
#model_dewpoint = filter_121(model_dewpoint, 20)

model_qv = interp1d(heights, ## g/kg -> kg/kg
                    sounding['mixing_ratio']/1000., kind='slinear')(zs[1:])
model_qv = prepend(sounding['mixing_ratio'].values[0]/1000., model_qv)
model_qv = soundings.filter_121(model_qv, 1)

model_dewpoint = soundings.calc_dewpoint(model_temperature,
                                         model_pressure,
                                         model_qv)

model_dir = interp1d(heights,
                     sounding['direction']*1.0, kind='slinear')(zs[1:])
model_dir = prepend(sounding['direction'].values[0], model_dir)

model_wspd = interp1d(heights,
                      sounding['sknt'], kind='slinear')(zs[1:])
model_wspd = prepend(sounding['sknt'].values[0], model_wspd)

dirs = np.pi*model_dir/180. # in radians
#dirs -= np.pi/2. # rotate 90 deg CCW
model_us = -model_wspd*np.sin(dirs)
model_vs = -model_wspd*np.cos(dirs)

model_data = {'temperature': model_temperature,
              'pressure': model_pressure,
              'dewpoint': model_dewpoint,
              'us': model_us,
              'vs': model_vs,
              'mixing_ratio': model_qv*1000.,
              'height': zs}
model_sounding = pd.DataFrame(model_data)

## MODIFICATIONS TO SOUNDING:
# 1) Set 0 velocity at surface 
#model_sounding.us.ix[0] = 0
#model_sounding.vs.ix[0] = 0

# 2) Reset the surface td/qv to match the layers above
model_sounding.mixing_ratio.ix[0] = model_sounding.mixing_ratio.ix[1]
model_sounding.dewpoint.ix[0] = model_sounding.dewpoint[1]

# 3) Raise the surface temperature by 5 degrees, and adiabatically raise
# the temp in the lowest model layers
T_sfc_heat = 2.
t_new = [model_sounding.temperature[0] + T_sfc_heat]
for i in xrange(1, len(zs)):
    if zs[i] > 0.5e3: break
    print i, zs[i]
    t_new.append(t_new[-1] - (zs[i]-zs[i-1])*9.81/1004.)
#print t_new
model_sounding.temperature.ix[:i-1] = t_new


# 4) Reduce water vapor above tropopause
'''
#print model_sounding[["pressure", "qv", "height"]]
above_trop = model_sounding["height"] > 11000
model_sounding.mixing_ratio[above_trop] = model_sounding.mixing_ratio[above_trop]*0.35
model_sounding.dewpoint[above_trop] = soundings.calc_dewpoint( \
                                        model_sounding.temperature[above_trop],
                                        model_sounding.pressure[above_trop],
                                        model_sounding.mixing_ratio[above_trop]*1e-3)
#print model_sounding[["pressure", "qv", "height"]]
'''

# 5) There's a discontinuity in the water vapor profile between
# 1.5 and 3.0 km. Smooth it out by fixing the qv mixing ratio to decrease
# linearly with height between these altitudes.
'''
layer = (model_sounding.height >= 1500.) & (model_sounding.height <= 3000.)
qv_layer = model_sounding.mixing_ratio[layer]

slope = (qv_layer.values[-1] - qv_layer.values[0])/(len(qv_layer)-1)
for i in xrange(1, len(qv_layer)-1):
    qv_layer.values[i] = qv_layer.values[0] + slope*i

model_sounding.mixing_ratio[layer] = qv_layer
model_sounding.dewpoint[layer] = soundings.calc_dewpoint( \
                                   model_sounding.temperature[layer],
                                   model_sounding.pressure[layer],
                                   model_sounding.mixing_ratio[layer]*1e-3)
'''

# 6) Try a totally new wind profile, following Rotunno et al (1988)
z = model_sounding.height.values*1e-3

u_old = model_sounding['us'].copy()
v_old = model_sounding['vs'].copy()

us = np.zeros_like(z)
u0 = 0 # m/s
z_cut = 6.0 # km
shear = 45. / z_cut # m/s / km
us[z <= z_cut] = u0 + z[z <= z_cut]*shear
us[z > z_cut] = us[z == z_cut]
model_sounding.us = us/0.51444 # calcualted in m/s -> convert to knots
model_sounding.vs[:] = 0.

print "shear = %1.4f s^1" % (shear/1000., )

###############################
## Finished with modifications!

soundings.plot_skewt(model_sounding, ax=ax1, first_index=0,
                        lift_parcel=True, plot_winds=True, diags=['CAPE'], 
                        T_color='k', Td_color='k', parcel_color='k',
                        barb_loc=48., barb_color='k', label_altitudes=True)
#ax1.set_xlim(-32)

## Correct model winds from knots -> m/s
model_sounding['us'] = model_sounding['us']*0.51444
model_sounding['vs'] = model_sounding['vs']*0.51444

plt.savefig("initial_sounding.pdf")
plt.savefig("initial_sounding.png")
```

#### Chemistry profiles

```python
## Set up profiles based on MOZART data
chem_data = pd.read_csv("mozart_04252011_18z.csv")
gas = np.zeros([nz, 17])

species = chem_data.columns
chem_pres = chem_data.pressure.values
model_pres = model_sounding['pressure'].values[:-1]

gases = ["o3", "co", "zco2", "ho", "ho2", "h2o2", "xno", "xno2", "xno3",
          "xn2o5", "hno3", "ch4", "ch2o", "ch3o2h", "so2", "h2so4", "dms"]
gas_map = { g:i for i, g in enumerate(gases) }

for s in gases:
    chem_spec = chem_data[s].values
    fit = interp1d(chem_pres, chem_spec, kind='slinear',
                   bounds_error=False, fill_value=chem_spec[-1])
    model_spec = fit(model_pres)
    
    gas[:, gas_map[s]] = model_spec
    
print "writing", namelist["file_cheminit"]


crm_io.write_initial_chem(namelist['file_cheminit'], gas)
_ = crm_io.read_initial_chem(namelist['file_cheminit'], nz)
```
