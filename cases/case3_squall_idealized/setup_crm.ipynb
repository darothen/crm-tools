## Setup CRM - 2D Version

---

This notebook is designed to quickly setup the initial conditions necessary to run the CRM. It will create the namelist files and some basic analysis plots of the soundings used to initialize the model. Then, it will copy the files in place as well as to the model run directory. The following set of cells will automate the setup and visual inspection of initial conditions for Chien's cloud-resolving model. For now, it is assumed that a homogeneous profile will be applied to the entire model domain.

__CASE__: Idealized 2D squall line from the [8th International Cloud Modeling Workshop](http://www.rap.ucar.edu/~gthompsn/workshop2012/case2/).

<img src="squall_skewt.gif">

```python
CASENAME = "squall"
CASEDIR = "."

USE_SKEWT = True

## Domain and sounding info
nx, ny, nz = 300, 1, 80

namelist = {
    ## Case settings
    "casename"      : CASENAME,
    #"homedir"       : "/Users/daniel/workspace/crm71_dev",
    "file_init"     : '%s_init' % CASENAME,
    "file_rktbl"    : 'rktable32.dat',
    "file_cheminit" : 'CHEMINIT_32_blank', 
            
    ## Time steps
    "dt"   : 5.0, # basic model time step (seconds)
    "dtau" : 0.5, # small time step for sound waves (seconds)
    "ds"   : 2.0, # time step for xxx (seconds)
    
    ## Resume control
    "nresume" : 0, # 0 = new run with one time field init data;
                   # 1 = start a run with two time field init data/apply disturbance
                   # 2 = resume a run with two time field init data
    
    ## Start point in step
    "nstart" : 1,
    
    ## Start time in 24 hour base:
    # The main program will convert to seconds, e.g.
    # start_in_s = 12.0 -> 43200 s (12 hours -> 43200 s)
    "start_in_s" : 12.0,
    
    ## Integration length:
    "nstop" : 6, # 24 if nresume == 0; 120 when nresume == 1
    "nunit" : 3600, # 3600: hour
                    # 60: minute
                    # dt: step
    
    ## Frequency of output data in step:
    "iax" : -1,
    
    ## Radiation calculation frequency:
    # example: 60 would be every 5 minutes if dt = 5 seconds
    # set to negative int or 0 to close radiation call
    "iradx" : 100, 
    
    ## Chemistry Package
    #"if_chem" : 1, # obsolete
    # Rxn calculation frequency in steps
    # (e.g. 6 would be every 30 seconds in 5 sec time step)
    "i_creact" : 6,
    
    ## Dimensions
    "ndims" : 2,   # numbe2r of dimensions
    "dx" : 2000.0, # x-width in meters
    "dy" : 2000.0, # y-width in meters
    "dz" : 250.0,  # height in meters
    
    ## Thickness of top damping layer
    # normally can be set to 5000(m)/dz(m)
    "kdamp" : 20,
    
    ## LBC parameters
    "eps" : 0.0, # relaxation coefficient
    "if_lbcdamp" : 0, # if > 0, tau = dx/C = n in seconds
    #"if_plbc" : 0, # periodic lbc option; obsolete
    
    ## Initial wind adjustments
    # Also controls the maximum 
    "u0shift" : 0.0, # m/s
    "v0shift" : 0.0, # m/s
    
    ## Minimum K for wind/scalar
    # reference: fkmin_u = 25.0, fkmin_s = 3.*fkmin_u
    "fkmin_u" : 25.0,
    "fkmin_s" : 75.0,
    
    ## Warm cloud?
    "nwarm" : 0, # 1 = warm, 0 = cold
    
    ## Go forward one step after smoothing?
    "nsmooth" : 1, # 1 = go one step forward, 0 = don't
    
    ## CCN and IN parameters
    "xnc0_c"  : 100.0e6,  # CCN nucleation constant C in CCN 1/kg
    "xnc0_k"  : 0.5,     # CCN nucleation constant k (0.7 for maritime, 0.5 for cont)
    "xn_ccn0" : 500.0e6, # sfc or max CCN concentration (1/kg)
    "xn_in0"  : 100.0e3, # sfc or max IN concentration (1/kg)
    
    ## Filter
    "if_filter" : -5, # if > 0, apply filter per n(dt) steps, e.g. 12
    
    ## Solar radiation
    "j_day" : 115, # day of the year from 1 to 365
    "ctr_lat" : 32.44, # latitude at center of model domain, from -90 to 90
    
    ## Initial perturbation
    "nbulb" : 2, # 1 = use bulb, 0 = don't
    "dptil" : 3.0, # max init perturbation of temperature in degree
    "dqv" : 0.9, # max init perturbation of water vapor mixing ratio in %-age
    "mx"  : 1,
    "mx1" : 1,
    "mx2" : 200,
    "ioe" : 2,
    "my"  : 1,
    "my1" : 1,
    "my2" : 1,
    "joe" : 2,
    "mz"  : 10,
    "mz1" : 5,
    "mz2" : 5,
    "koe" : 2,
}
```

```python
## Write namelist

nl_order = ["dt", "dtau", "ds", "nresume", "nstart", "start_in_s", "nstop",
            "nunit", "iax", "iradx", "i_creact", "ndims", "dx", "dy", "dz",
            "kdamp", "eps", "if_lbcdamp", "u0shift", "v0shift", "fkmin_u",
            "fkmin_s", "nwarm", "nsmooth", "xnc0_c", "xnc0_k", "xn_ccn0",
            "xn_in0", "if_filter", "nbulb", "dptil", "dqv", "mx", "mx1", 
            "mx2", "ioe", "my", "my1", "my2", "joe", "mz", "mz1", "mz2",
            "koe", "j_day", "ctr_lat", "casename", "file_init",
            "file_rktbl", "file_cheminit"]

namelist_fn = "cm%1dd.nml" % namelist['ndims']
with open(namelist_fn, "w") as f:
    f.write("$cm_in \n")
    for key in nl_order:
        value = namelist[key]
        f.write("   %-13s = %r \n" % (key, value))
    f.write("$END \n")
print namelist_fn
```

####Thermodynamic soundings/profiles

```python
P_sfc = 978.0 # hPa
theta_sfc = 303.068 # K
qv_sfc = 21.077 # g/kg
T_sfc = theta_sfc*((P_sfc/1000.)**(soundings.Rd/soundings.cp)) - 273.15

## Read the interpolated temperature Skew-T
skewt_data = np.loadtxt("t_p.txt", delimiter=",")
skewt_data = pd.DataFrame(skewt_data, index=range(1, len(skewt_data)+1))
skewt_data.columns = ["T", "P"]

# Read in the dewpoints
skewt_td = np.loadtxt("td_p.txt", delimiter=",")
skewt_td = pd.DataFrame(skewt_td, index=range(1, len(skewt_td)+1))
skewt_td.columns = ["Td", "P"]

# map dewpoints to temperatures from other interpolated file
print " T-DATA", skewt_data.P.min(), skewt_data.P.max()
print "Td-DATA", skewt_td.P.min(), skewt_td.P.max()
x = skewt_data.P.values[::-1]
y = skewt_data["T"].values[::-1]
xp = skewt_td.P.values[::-1]
print x[:3], x[-3:]
print xp[:3], xp[-3:]
mapped_ts = interp1d(x, y)(xp)[::-1]

skewt_td["T"] = mapped_ts

# Calculate qv, theta, theta_e based on T, P, Td
from scipy.optimize import bisect
qvs, thetas, theta_ms, theta_es = [], [], [], []
for _, data in skewt_td.iterrows():
    f_qv = lambda r: soundings.calc_dewpoint(data["T"], data["P"], r) - data["Td"]
    qv = bisect(f_qv, 1e-6, 30)
    qvs.append(qv)
    
    thetas.append(soundings.calc_theta(data["T"], data["P"]))
    theta_ms.append(soundings.calc_theta(data["T"], data["P"], qv))
    theta_es.append(soundings.calc_theta_e(data["T"], data["P"], qv))
    
skewt_td["QV"] = qvs # kg/kg
skewt_td["THETA"] = thetas # K
skewt_td["THETA_M"] = theta_ms # K
skewt_td["THETA_E"] = theta_es # K

# Append sfc values
sfc = pd.DataFrame({"P": P_sfc, "THETA": theta_sfc, "QV": qv_sfc/1e3, "T": T_sfc}, 
                   index=[0,])
skewt_td = sfc.append(skewt_td)
data_len = len(skewt_td)
print skewt_td.head()

sounding_data = { 'temperature': skewt_td['T'], 'dewpoint': skewt_td['Td'],
                  'pressure': skewt_td['P'], 
                  'mixing_ratio': skewt_td['QV']/1e3,
                  'height': np.zeros(data_len) }
snd = pd.DataFrame(sounding_data)

#soundings.plot_skewt(snd)

#plt.figure()
#skewt_td[["THETA", "THETA_M", "THETA_E"]].plot()

#plt.figure()
#plt.plot(skewt_td.THETA - skewt_td.THETA_M)
```

```python
## Read the providing sounding data
cols = ["HEIGHT", "THETA", "QV", "U", "V"]
units = ["m", "K", "g/kg", "m/s", "m/s"]

height_ref = np.array([
    16000., 16500., 17000., 17500., 18000., 18500.,
    19000., 19500., 20000., 20500., 21000., 21500., 22000., 22500., 23000.,
    23500., 24000., 24500.,])
pres_ref = np.array([
    105.82300568, 97.45291138,
    89.7336731, 82.49432373, 75.89122772, 69.88571167, 64.33828735,
    59.27000046, 54.67272568, 50.41570282, 46.49965286, 42.92856979,
    39.63215637, 36.58787537, 33.78852463, 31.25, 28.85444832,
    26.64360619])

snd_data = np.loadtxt("initial_sounding.txt")
snd_data = pd.DataFrame(snd_data, index=range(1, len(snd_data)+1), 
                        columns=cols)
sfc = pd.DataFrame({ "HEIGHT": 0., "THETA": theta_sfc, "QV": qv_sfc, 
                     "U": -12., "V": 0 }, index=[0,])
snd_data = sfc.append(snd_data)

## Use the mapping of THETA -> HEIGHT from the sounding data to analyze the 
## height corresponding to the interpolated skew-t data
x = snd_data.THETA.values
y = snd_data.HEIGHT.values
xp = skewt_td.THETA.values
print x[:5]
print y[:5]
print xp[:5]
mapped_height = interp1d(x, y, bounds_error=True)(xp)

## Use the mapping of THETA -> P from the interpolated skewt-data to
## analyze the height corresponding to the sounding data
x = skewt_td.THETA.values
y = skewt_td.P.values
xp = snd_data.THETA.values
mapped_pres = interp1d(x, y, bounds_error=False)(xp)
```

Use the `mapped_height` with the __skew-t data__ to create an input dataset for the CRM.

```python
if USE_SKEWT:
    dx, dy, dz = namelist['dx'], namelist['dy'], namelist['dz']

    fig, ax1 = subplots(1,1, figsize=(7, 7), 
                        subplot_kw={'axisbg': 'w'} )

    soundings.plot_skewt(snd, ax=ax1, lift_parcel=False,
                         diags=[], title="",
                         T_color='grey', Td_color='grey', parcel_color='grey',
                         barb_loc=42., barb_color='grey',
                         T_left=-32.)

    zs = np.linspace(0, dz*nz, nz+1)[:-1]# + 125.0
    
    new_skewt = skewt_td.copy()

    new_skewt['HEIGHT'] = mapped_height

    ## Map temperatures in the stratosphere, z > 16km
    ## Map pressures in stratosphere, z > 16km, P < 100 mb
    zs_strat = zs[zs > 16e3]
    tropop_temp = new_skewt['T'].values[-1]
    dT_dz = (211.114470027 - 205.27030149) / (19500.0  - 16000.0)
    p_by_z = interp1d(height_ref, pres_ref, bounds_error=False)
    pressures, temps = [], []
    for i, z in enumerate(zs_strat):
        temps.append(tropop_temp + dT_dz*(dz*(i+1)))
        pressures.append(p_by_z(z))
    pressures = np.squeeze(np.array(pressures))
    qvs = np.ones_like(temps)*new_skewt.QV.values[-1]

    new_heights = np.concatenate((new_skewt.HEIGHT, zs_strat))
    new_pressures = np.concatenate((new_skewt.P, pressures))
    new_temperatures = np.concatenate((new_skewt['T'], temps))
    new_qvs = np.concatenate((new_skewt.QV, qvs))

    #print len(new_heights), len(new_pressures), len(new_temperatures)
    new_rows = pd.DataFrame( {'HEIGHT': new_heights,
                              'QV': new_qvs,
                              'P': new_pressures,
                              'T': new_temperatures} )
    new_skewt = new_skewt.append(new_rows)

    print new_skewt.head()
    print new_skewt.tail()

    ###

    heights = new_heights            # m 
    pressures = new_pressures        # hPa
    temperatures = new_temperatures  # C
    qv = new_qvs                     # kg/kg

    ## PRESSURE 
    model_pressure = interp1d(heights, pressures, bounds_error=False)(zs) # hPa

    ## TEMPERATURE
    model_temperature = interp1d(heights, temperatures, bounds_error=False)(zs) + 273.15 # K

    ## QV
    model_qv = interp1d(heights, qv, bounds_error=False)(zs)*1e3 # g/kg

    ## U
    model_us = np.zeros_like(model_qv)
    amp = 12.     # m/s
    depth = 2500. # m
    shear = amp/depth # s^-1
    for i in xrange(len(model_us)):
        if zs[i] > depth: break
        model_us[i] = -amp + shear*i*dz

    ## V
    model_vs = np.zeros_like(model_us) # m/s
    
    ## CORRECTIONS BASED ON REFERENCE SOUNDING
    '''
    theta_ref = [304.6479, 305.1141, 305.3860, 306.8239]
    qv_ref = [20.6819, 20.3964, 20.2731, 18.7877]
    for i in [2, 1, 0]:
        p = model_pressure[i+1]
        th = theta_ref[i+1]
        kappa = soundings.Rd/soundings.cp
        dp_dz = (9.81/soundings.Rd/th)*(1000.**kappa)*(p**(1.-kappa))
        print i, dp_dz
        model_pressure[i] = p + dp_dz*dz
    model_qv[:3] = qv_ref[:3]
    model_temperature[:3] = theta_ref[:3]*((model_pressure[:3]/1000.)**kappa)
    '''
        

    ## DEWPOINT
    model_dewpoint = soundings.calc_dewpoint(model_temperature,
                                             model_pressure,
                                             model_qv/1e3)

    ######################

    model_data = {'temperature': model_temperature - 273.15,
                  'pressure': model_pressure,
                  'dewpoint': model_dewpoint,
                  'us': model_us/0.51444, # m/s -> knots for plotting
                  'vs': model_vs/0.51444,
                  'mixing_ratio': model_qv,
                  'height': zs}
    model_sounding = pd.DataFrame(model_data)

    print model_sounding.head()
    print model_sounding.tail()

    soundings.plot_skewt(model_sounding, ax=ax1, 
                         T_color='k', Td_color='k', parcel_color='k',
                         barb_loc=48., barb_color='k', label_altitudes=True)
```

Use the `mapped_pres` with the __initial sounding data__ to create an input dataset for the CRM.

```python
if USE_SKEWT: 
    print "not used"
else:
    dx, dy, dz = namelist['dx'], namelist['dy'], namelist['dz']

    fig, ax1 = subplots(1,1, figsize=(7, 7), 
                        subplot_kw={'axisbg': 'w'} )

    soundings.plot_skewt(snd, ax=ax1, lift_parcel=False,
                         diags=[], title="",
                         T_color='grey', Td_color='grey', parcel_color='grey',
                         barb_loc=42., barb_color='grey',
                         T_left=-32.)

    zs = np.linspace(0, dz*nz, nz+1)[:-1]

    ## We need to flesh out the pressure values for the stratospheric altitudes
    ## in the sounding data. To do this, we will use the reference data
    ## from above to fill in where there are NaNs in the mapped pressure
    heights = snd_data['HEIGHT'].values # m
    pressures = mapped_pres
    p_by_z = interp1d(height_ref, pres_ref, bounds_error=False)
    for i, (p, z) in enumerate(zip(pressures, heights)):
        if np.isnan(p): 
            print i, p, z, 
            print p_by_z(z)
            pressures[i] = p_by_z(z)

    thetas = snd_data['THETA']
    temperatures = thetas*((pressures/1000.)**(soundings.Rd/soundings.cp))
    qv = snd_data['QV'].values # g/kg
    us = snd_data['U'].values # m/s
    vs = snd_data['V'].values # m/s

    ## PRESSURE 
    model_pressure = interp1d(heights, pressures)(zs) # hPa

    ## THETA
    model_theta = interp1d(heights, thetas)(zs) # K

    ## TEMPERATURE
    #model_temperature = interp1d(heights, temperatures)(zs) # K
    model_theta*((model_pressure/1000.)**(soundings.Rd/soundings.cp))

    ## QV
    model_qv = interp1d(heights, qv)(zs) # g/kg

    ## U
    model_us = interp1d(heights, us)(zs) # m/s

    ## V
    model_vs = interp1d(heights, vs)(zs) # m/s

    ## DEWPOINT
    model_dewpoint = soundings.calc_dewpoint(model_temperature,
                                             model_pressure,
                                             model_qv/1e3)

    ######################

    model_data = {'temperature': model_temperature - 273.15,
                  'pressure': model_pressure,
                  'dewpoint': model_dewpoint,
                  'us': model_us/0.51444, # m/s -> knots for plotting
                  'vs': model_vs/0.51444,
                  'mixing_ratio': model_qv,
                  'height': zs}
    model_sounding = pd.DataFrame(model_data)

    soundings.plot_skewt(model_sounding, ax=ax1, 
                         T_color='k', Td_color='k', parcel_color='k',
                         barb_loc=48., barb_color='k', label_altitudes=True)
```
