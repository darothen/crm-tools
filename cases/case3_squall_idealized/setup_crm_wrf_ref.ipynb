## Setup CRM - WRF Reference (2D) Version

---

Similar to `setup_wrf_wrf`, this notebook uses a given ASCII sounding to generate initial conditions for the CRM. However, it goes a step further by using the WRF (v3.6) model's routine for diagnosing and analyzing the model pressure from the given $Z$, $\theta$, and $q_v$.

__CASE__: Idealized 2D squall line from the [8th International Cloud Modeling Workshop](http://www.rap.ucar.edu/~gthompsn/workshop2012/case2/).

```python
CASENAME = "squall_wrf_rotated"
CASEDIR = "."

## Domain and sounding info
nx, ny, nz = 400, 1, 80

namelist = {
    ## Case settings
    "casename"      : CASENAME,
    #"homedir"       : "/Users/daniel/workspace/crm71_dev",
    "file_init"     : '%s_init' % CASENAME,
    "file_rktbl"    : 'rktable32.dat',
    "file_cheminit" : 'CHEMINIT_32_blank', 
            
    ## Time steps
    "dt"   : 2.0, # basic model time step (seconds)
    "dtau" : 0.2, # small time step for sound waves (seconds)
    "ds"   : 2.0, # time step for xxx (seconds)
    
    ## Resume control
    "nresume" : 0, # 0 = new run with one time field init data;
                   # 1 = start a run with two time field init data/apply disturbance
                   # 2 = resume a run with two time field init data
    
    ## Start point in step
    "nstart" : 1,
    
    ## Start time in 24 hour base:
    # The main program will convert to seconds, e.g.
    # start_in_s = 12.0 -> 43200 s (12 hours -> 43200 s)
    "start_in_s" : 12.0,
    
    ## Integration length:
    "nstop" : 9, # 24 if nresume == 0; 120 when nresume == 1
    "nunit" : 3600, # 3600: hour
                    # 60: minute
                    # dt: step
    
    ## Frequency of output data in step:
    "iax" : -1,
    
    ## Radiation calculation frequency:
    # example: 60 would be every 5 minutes if dt = 5 seconds
    # set to negative int or 0 to close radiation call
    "iradx" : 100, 
    
    ## Chemistry Package
    #"if_chem" : 1, # obsolete
    # Rxn calculation frequency in steps
    # (e.g. 6 would be every 30 seconds in 5 sec time step)
    "i_creact" : 6,
    
    ## Dimensions
    "ndims" : 2,   # numbe2r of dimensions
    "dx" : 1000.0, # x-width in meters
    "dy" : 1000.0, # y-width in meters
    "dz" : 250.0,  # height in meters
    
    ## Thickness of top damping layer
    # normally can be set to 5000(m)/dz(m)
    "kdamp" : 20,
    
    ## LBC parameters
    "eps" : 0.0, # relaxation coefficient
    "if_lbcdamp" : 0, # if > 0, tau = dx/C = n in seconds
    #"if_plbc" : 0, # periodic lbc option; obsolete
    
    ## Initial wind adjustments
    # Also controls the maximum 
    "u0shift" : 3.0, # m/s
    "v0shift" : 0.0, # m/s
    
    ## Minimum K for wind/scalar
    # reference: fkmin_u = 25.0, fkmin_s = 3.*fkmin_u
    "fkmin_u" : 25.0,
    "fkmin_s" : 75.0,
    
    ## Warm cloud?
    "nwarm" : 0, # 1 = warm, 0 = cold
    
    ## Go forward one step after smoothing?
    "nsmooth" : 1, # 1 = go one step forward, 0 = don't
    
    ## CCN and IN parameters
    "xnc0_c"  : 100.0e6,  # CCN nucleation constant C in CCN 1/kg
    "xnc0_k"  : 0.5,     # CCN nucleation constant k (0.7 for maritime, 0.5 for cont)
    "xn_ccn0" : 500.0e6, # sfc or max CCN concentration (1/kg)
    "xn_in0"  : 100.0e3, # sfc or max IN concentration (1/kg)
    
    ## Filter
    "if_filter" : -5, # if > 0, apply filter per n(dt) steps, e.g. 12
    
    ## Solar radiation
    "j_day" : 115, # day of the year from 1 to 365
    "ctr_lat" : 32.44, # latitude at center of model domain, from -90 to 90
    
    ## Initial perturbation
    "nbulb" : 2, # 1 = use bulb, 0 = don't
    "dptil" : 3.0, # max init perturbation of temperature in degree
    "dqv" : 0.9, # max init perturbation of water vapor mixing ratio in %-age
    "mx"  : 1,
    "mx1" : 125,
    "mx2" : 150,
    "ioe" : 2,
    "my"  : 1,
    "my1" : 1,
    "my2" : 1,
    "joe" : 2,
    "mz"  : 18,
    "mz1" : 1,
    "mz2" : 1,
    "koe" : 2,
}
```

```python
## Write namelist

nl_order = ["dt", "dtau", "ds", "nresume", "nstart", "start_in_s", "nstop",
            "nunit", "iax", "iradx", "i_creact", "ndims", "dx", "dy", "dz",
            "kdamp", "eps", "if_lbcdamp", "u0shift", "v0shift", "fkmin_u",
            "fkmin_s", "nwarm", "nsmooth", "xnc0_c", "xnc0_k", "xn_ccn0",
            "xn_in0", "if_filter", "nbulb", "dptil", "dqv", "mx", "mx1", 
            "mx2", "ioe", "my", "my1", "my2", "joe", "mz", "mz1", "mz2",
            "koe", "j_day", "ctr_lat", "casename", "file_init",
            "file_rktbl", "file_cheminit"]

namelist_fn = "cm%1dd.nml" % namelist['ndims']
with open(namelist_fn, "w") as f:
    f.write("$cm_in \n")
    for key in nl_order:
        value = namelist[key]
        f.write("   %-13s = %r \n" % (key, value))
    f.write("$END \n")
print namelist_fn
```

####Thermodynamic soundings/profiles

```python
P_sfc = 978.0 # hPa
theta_sfc = 303.068 # K
qv_sfc = 21.077 # g/kg
u_sfc = -12. # m/s
v_sfc = 0.   # m/s
kappa = soundings.Rd/soundings.cp
T_sfc = theta_sfc*((P_sfc/1000.)**kappa) - 273.15
theta_v_sfc = theta_sfc*(1. + 0.61*qv_sfc/1e3)

## Read the reference sounding ASCII file using the WRF read routine
n = 100
dry = False
zk, p, pdry, theta, rho, u, v, qv, nl = wrf_io.get_sounding("input_sounding", 
                                                            dry, n)
zk = zk[:nl]
p = p[:nl]
pdry = pdry[:nl]
theta = theta[:nl]
rho = rho[:nl]
u = u[:nl]
v = v[:nl]
qv = qv[:nl]

## Prepare data for plotting on Skew-T
ps = [P_sfc*100, ] + p.tolist() # Pa
qvs = [qv_sfc*1e-3, ] + qv.tolist() # kg/kg

temps = theta*((p/1e5)**(287./1004.))
ts = [T_sfc+273.15, ] + temps.tolist()
heights = [0., ] + zk.tolist()
us = [u_sfc, ] + u.tolist()
vs = [v_sfc, ] + v.tolist()

####
dz = namelist['dz']
zs = np.linspace(0, dz*nz, nz+1)[:-1]# + 125.0

## PRESSURE 
model_pressure = interp1d(heights, ps, bounds_error=False)(zs) # Pa

## TEMPERATURE
model_temperature = interp1d(heights, ts,
                             bounds_error=False)(zs) # K

## QV
model_qv = interp1d(heights, qvs, bounds_error=False)(zs) # g/kg

## U
model_us = interp1d(heights, us, bounds_error=False)(zs) # m/s

## V
model_vs = interp1d(heights, vs, bounds_error=False)(zs) # m/s


## DEWPOINT
model_dewpoint = soundings.calc_dewpoint(model_temperature - 273.15,
                                         model_pressure/100.,
                                         model_qv)

# rotate winds
model_vs = model_us[:]
model_us = np.zeros_like(model_vs)

######################

model_data = {'temperature': model_temperature - 273.15,
              'pressure': model_pressure/100.,
              'dewpoint': model_dewpoint,
              'us': model_us/0.51444, # m/s -> knots for plotting
              'vs': model_vs/0.51444,
              'mixing_ratio': model_qv*1e3,
              'height': zs}
model_sounding = pd.DataFrame(model_data)

print model_sounding.head()
print model_sounding.tail()

ax, _ = soundings.plot_skewt(model_sounding, lift_parcel=True, 
                             diags=['CAPE'], plot_winds=True,
                             T_color='k', Td_color='k', parcel_color='k',
                             barb_loc=48., barb_color='k', 
                             label_altitudes=True)
```

```python
from model_util import diag_plots

# compute N^2, using dry atm (N^2 = g/theta * d_theta/d_z )
pres = model_sounding['pressure'].values*100 # Pa
temp = model_sounding['temperature'].values + 273.15 # K

print pres.min(), pres.max()
print temp.min(), temp.max()

diag_plots.gravity_wave_plot(zs, pres, temp)
```

```python
## Hydrostatic balance

qv = model_sounding['mixing_ratio'].values/1000.
diag_plots.hydrostatic_plot(zs, pres, temp, qv)
```

```python
print "writing", namelist['file_init']

print model_sounding.head()

crm_io.write_initial_profile(namelist['file_init'],
                             model_sounding['pressure'].values,
                             model_sounding['height'].values,
                             model_sounding['temperature'].values+273.15, # C -> K
                             model_sounding['us'].values*0.51444,
                             model_sounding['vs'].values*0.51444,
                             model_sounding['mixing_ratio'].values/1000.)
_ = crm_io.read_initial_profile(namelist['file_init'], nz, annika=False)
```
