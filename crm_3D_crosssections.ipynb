3D Cross Sections of CRM Output
===============================

Following the analysis by [Kuhlman et al, 2006](http://journals.ametsoc.org/doi/abs/10.1175/MWR3217.1), this script generates a sequence of cross-sectional images, specifically Figures 8 (d-f) and Figure 11. It can be tweaked to generate these plots for different height levels, timestamps, or cross-section planar cuts. Note that only purely horizontal cuts can be made at this time (cuts on an angle will be implemented later).

---

Read in the CRM netCDF file to analyze using `iris`

```python
filename = "supercell_master.nc"

cubes = iris.load(filename)
data = {}
for c in cubes: 
    data[c.var_name] = c
ex_scalar = data['QC']
time = ex_scalar.coord(var_name="time")
```

Before doing anything, we need to re-map the velocity fields to the standard grid

```python
xc  = ex_scalar.coord(var_name='xc')
xcu = data['U'].coord(var_name='xcu')
yc  = ex_scalar.coord(var_name='yc')
ycv = data['V'].coord(var_name='ycv')
zc  = ex_scalar.coord(var_name='zc')
zcw = data['W'].coord(var_name='zcw')

print "W"
w_new = iris.analysis.interpolate.linear(data['W'], 
                                         [(zcw.long_name, zc.points), ])
w_new.remove_coord(zcw.long_name)
w_new.add_dim_coord(zc, 1)
data['W'] = w_new

print "V" 
v_new = iris.analysis.interpolate.linear(data['V'], 
                                         [(ycv.long_name, yc.points), ])
v_new.remove_coord(ycv.long_name)
v_new.add_dim_coord(yc, 2)
data['V'] = v_new

print "U"
u_new = iris.analysis.interpolate.linear(data['U'], 
                                         [(xcu.long_name, xc.points), ])
u_new.remove_coord(xcu.long_name)
u_new.add_dim_coord(xc, 3)
data['U'] = u_new
```

To check that everythinhg worked out okay, let's plot the sample from Figure 8.d, using the time t=116 minutes and z = 6.8 kilometers. Since we don't have access to radar reflectivities with the raw data, we will plot the hoirzontal wind vectors and an outline of the cloud using the contour QC=0.01 kg/kg.

```python
import iris.plot as iplt
import iris.quickplot as qplt

## Setup the constraints
alt = iris.Constraint(**{zc.name(): 6875.})
ts = iris.Constraint(time=144*60)
x_sub = iris.Constraint(**{xc.name(): lambda x: x > 60e3})
y_sub = iris.Constraint(**{yc.name(): lambda y: y > 60e3})

v = data['V'].extract(alt & ts)
u = data['U'].extract(alt & ts)
qc = data['QC'].extract(alt & ts)

fig = plt.figure()
ax = fig.add_subplot(111, axisbg='w')
qplt.contour(qc, coords=[xc.name(), yc.name()],
             colors='grey', levels=[1e-5, ],
             alpha=0.8, linewidths=2.5)
xi, xe = 85e3, 135e3
yi, ye = 20e3, 70e3
ax.set_xlim(xi, xe)
ax.set_ylim(yi, ye)


sl = 4
ax.quiver(xc.points[::sl], yc.points[::sl], 
          u.data[::sl, ::sl], v.data[::sl, ::sl],
          units='inches', scale=100)


lbl_dict = { 'weight': 'bold', 'family': 'sans', 'size': 16 }

## Vertical slice
CD_y = [25e3, 65e3]
CD_x = 100e3
ax.plot([CD_x, CD_x], CD_y, 'ok', linestyle='solid', lw=2)
ax.text(CD_x+500, CD_y[0], "C", fontdict=lbl_dict)
ax.text(CD_x+500, CD_y[1], "D", fontdict=lbl_dict)

## Oblique slice
AB_x = [90e3, 135e3]
AB_y = [45e3, 35e3]

ax.plot(AB_x, AB_y, 'ok', linestyle='solid', lw=2)
ax.text(AB_x[0]+500, AB_y[0], "A", fontdict=lbl_dict)
ax.text(AB_x[1]+500, AB_y[1], "B", fontdict=lbl_dict)
```

```python
x_sub = iris.Constraint(**{xc.name(): CD_x+500})

u = data['U'].extract(ts & x_sub)
v = data['V'].extract(ts & x_sub)
w = data['W'].extract(ts & x_sub)
qc = data['QC'].extract(ts & x_sub)
qr = data['QR'].extract(ts & x_sub)
qg = data['QG'].extract(ts & x_sub)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, axisbg='w')
qplt.contour(qc, coords=[yc.name(), zc.name()],
             colors='grey', alpha=0.8,
             levels=[1e-5, ], linewidths=2.5,)
qplt.contour(qr, coords=[yc.name(), zc.name()],
              cmap=cm.jet)
ax.set_xlim(*CD_y)
ax.set_ylim(0, 16e3)

sl = 3
ax.quiver(yc.points[::sl], zc.points[::sl], 
          v.data[::sl, ::sl], w.data[::sl, ::sl],
          units='inches', scale=75, )
```
